<h2>DENGUE CASES TREND FROM 2016 - 2020</h2> 

```python
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
dengue['Month'] = pd.Categorical(dengue['Month'], categories=month_order, ordered=True)

monthly_cases = dengue.groupby(['Month', 'Year'])["Dengue_Cases"].sum()

monthly_cases.unstack().plot(kind='line')
plt.gcf().set_size_inches(12,8)
plt.grid(True)
plt.xticks(range(0,12), month_order)
plt.show()
```

<h2>Insight #1</h2>
<h3>If we were to analyze the graph we can see that from the month of june is where usually the cases of dengue increases, because from june to november is the rainy season of the philippines.</h3>
